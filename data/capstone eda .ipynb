```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the demographics dataframe

demographics = pd.read_csv('./demographics.csv')
demographics
```

```python
#look at column names
demographics.columns
```

```python
#check for missing values
df_list = [demographics, outcome, survey]
df_list_names = ["demographics", "outcome", "survey"]

for i in range(len(df_list)):
    df = df_list[i]
    name = df_list_names[i]
    null_counts = df.isnull().sum()
    if null_counts.any():
        print(f"{name} has null values:")
        print(null_counts)
    else:
        print(f"{name} doesn't have null values")
```

```python
##keeping the missing values in outcome dataframe because the empty values represents that the student did not manage to secure an offer
##keeping the missing values in survey dataframe because the students might have important data after merge with the other dataframe
```

```python
#the columns contain null values in the demographics dataframe will be removed as the whole columns are empty
demographics = demographics.drop(['items', 'Status.2','Course ID (4)'], axis=1) # drop the columns
demographics = demographics.drop(['Preferred Name','Original Course ID', 'Status.1','Status.3','Status.4','Course ID (2)', 'Course ID (3)','Course ID (5)'], axis=1) # drop the columns
demographics = demographics.drop(['Alternative Email','Emergency Contact', 'Special Learning Needs','What is your native language'], axis=1) # drop the columns
demographics = demographics.drop(["Have you participated in any similar programme? For eg: Penjana, MySTEP, Protege, SL1M, etc (if yes please state which programme you're join)", "I have a working computer or laptop that will be available to me throughout the duration of my course and I have a working internet connection that can handle video streaming."], axis=1) # drop the columns
demographics = demographics.drop(['Current/Last employment date'], axis=1) # drop the columns 
demographics = demographics.drop(['Job position Category'], axis=1) # drop the columns 
```

```python
#check for the remaining missing values in the demographic dataframe 
demographics.isnull().sum()

##keeping all other columns with missing values because other information of the student might be important, and responses are for question 'Please specify if choose others', so it's natural that they are empty 
```

```python
##filling the null values with mode has bigger descrapancies in mean values compared to filling null values with mean
## so will proceed with filling null values with mean
```

```python
#missing values in survey dataframe

#columns that need to be filled
columns_tofill = ['NPS','Satisfaction scores','value-time-spent', 'Content', 'Instructor', 'Experience']

#mean scores by course
mean_by_course = survey.groupby('Course ID ')[columns_tofill].mean().round(1)

#fill in the null values 
for column in columns_tofill:
    survey[column] = survey.apply(lambda x: mean_by_course.loc[x['Course ID '], column] if pd.isnull(x[column]) else x[column], axis=1)

    
#check if all null values are filled
survey.isnull().value_counts()
```

```python
#joining 3 dataframes together 
combined_df= demographics.merge(survey,on='Email').merge(outcome,on='Email')
combined_df
```

```python
#remove the repetitive columns like full name and course ids
combined_df = combined_df.drop(['Full Name_y','Course ID _x','Full Name','Course ID _y'], axis=1) # drop the columns 
combined_df.columns
```

```python
#phone number should not be a float, so will convert it to int 
#there are some missing info in the column, so will fill the missing info with 0
combined_df['Phone'] = combined_df['Phone'].fillna(0)
combined_df['Phone'] = pd.to_numeric(combined_df['Phone'], errors='coerce')

combined_df['Phone'] = combined_df['Phone'].astype(int)
combined_df.dtypes
```

```python
#convert attendance rate from obj to int
combined_df['Attendance rate'] = combined_df['Attendance rate'].astype(str)
combined_df['Attendance rate'] = combined_df['Attendance rate'].str.rstrip('%').astype(float)
combined_df['Attendance rate']
```

# Learners Performance 

```python
#overall completion status
combined_df['Status'].value_counts()
```

```python
#breakdown of completion status by course
status_by_course= combined_df.groupby(['Grouped_Course', 'Status']).size()
status_by_course
```

```python
completion_rate = ((status_by_course / status_by_course.groupby('Grouped_Course').sum())* 100).round(2) 
completion_rate
```

```python
#find out the top scorer for each course
#CCST and CITM are not included as the rubric for completion is either completed or failed

courses_with_scores = ['DABC', 'DMBC', 'PPBC', 'PMBC', 'UXBC', 'DSBC']
top_performers_df = pd.DataFrame(columns=['Grouped_Course', 'Full Name_x', 'Final project score (Fail/ Pass/ scores)', 'Attendance rate', 'Placement Status '])

for course in courses_with_scores:
    course_df = combined_df[combined_df['Grouped_Course'] == course]
   
    # Find the maximum score for the current course
    max_score = course_df['Final project score (Fail/ Pass/ scores)'].max()
    
    # Find the row(s) with the maximum score for the current course
    max_score_rows = course_df[course_df['Final project score (Fail/ Pass/ scores)'] == max_score]

    top_performers_df = pd.concat([top_performers_df, max_score_rows])
```

# Learners Satisfation

```python
#let's see student's scores (from survey) based on course

combined_df['Satisfaction scores'].value_counts(), combined_df['value-time-spent'].value_counts(), combined_df['Content'].value_counts(),combined_df['Instructor'].value_counts(),combined_df['Experience'].value_counts()

#columns to count
columns_tocount = ['Satisfaction scores','value-time-spent', 'Content', 'Instructor', 'Experience']

#mean scores by course
mean_by_course = combined_df.groupby('Grouped_Course')[columns_tocount].mean().round(1)
mean_by_course = pd.DataFrame(mean_by_course)
mean_by_course

##DMBC and UXBC learners rated highest satisfaction score, wheras CSCT learners has the lowest satisfaction scores
```

```python
ax= mean_by_course.plot(kind='bar', stacked=True, figsize=(10, 4))

for container in ax.containers:
    ax.bar_label(container, label_type='center', color='white', fontweight='bold')
    
plt.title('Satisfaction Scores by Course')
plt.xlabel('Course')
plt.ylabel('Scores')
plt.xticks(rotation = 0)
plt.legend(title='Scores')
ax.legend(bbox_to_anchor=(1.0, 1.05))

plt.show()
```

```python
#calculate NPS for each course 

# categorize responses into promoters, passives, and detractors
promoters = combined_df[combined_df['NPS'] >= 9].groupby('Grouped_Course').size()
passives = combined_df[(combined_df['NPS'] >= 7) & (combined_df['NPS'] < 9)].groupby('Grouped_Course').size()
detractors = combined_df[combined_df['NPS'] <= 6].groupby('Grouped_Course').size()

#fill in detractor count as 0
all_courses = combined_df['Grouped_Course'].unique()
detractors = detractors.reindex(all_courses, fill_value=0)



# calculate the percentage of promoters, passives, and detractors 
total_responses = promoters + passives + detractors
promoter_percentage = promoters / total_responses * 100
detractor_percentage = detractors / total_responses * 100


# calculate the NPS for each course
nps_by_course = (promoter_percentage - detractor_percentage).round(1)


print(nps_by_course)

##PMBC has the highest NPS score; CCST has the lowest NPS score
```

```python
ax2 = nps_by_course.plot(kind='bar', figsize=(10, 4))

for container in ax2.containers:
    ax2.bar_label(container, label_type='edge', fontweight='bold')

plt.title('NPS by Course')
plt.xlabel('Course')
plt.ylabel('Scores')
plt.xticks(rotation=0)

ax2.set_ylim([0, 100])

plt.show()

##looks like the courses do not invlove data analysis score higher in NPS
```
