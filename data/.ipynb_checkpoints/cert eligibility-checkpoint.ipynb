```python
# Define the eligibility criteria
def passing_score(score):
    if isinstance(score, str):
        # Check if the value is 'completed'
        if score.lower().strip() == 'completed':
            return True  # Not a passing score
        else:
            try:
                score = float(score.rstrip('%'))
                return score >= 55
            except ValueError:
                return False  # Non-numeric string
    elif isinstance(score, (int, float)):
        return score >= 55
    else:
        return False
    
combined_data['Placement Status '] = combined_data['Placement Status '].str.strip( )

completed_finalproject = combined_data['Final project score (Fail/ Pass/ scores)'].apply(passing_score)
achieve_attendance = combined_data['Attendance rate'] >= 80
placement_status = combined_data['Placement Status '].isin(['Permanent', 'Contract', 'Self-employment'])

# Combine the criteria to determine eligibility
combined_data['Cert Eligibility'] = completed_finalproject & achieve_attendance & placement_status

```
